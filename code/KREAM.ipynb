```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
```

```python
browser = webdriver.Chrome()
```

```python
url = "https://kream.co.kr/search?tab=49&gender=men&sort=popularity_without_trading"
browser.get(url)
time.sleep(1)
```

```python
soup = BeautifulSoup(browser.page_source, 'html.parser')
```

```python
from selenium.webdriver.common.keys import Keys

soup = BeautifulSoup(browser.page_source, "html.parser")
total_list = soup.select('div.filter_result > span.title')[0].text.split(' ')
total = total_list[1]  

prev_length = None  
unchanged_count = 0  

while True:
    body = browser.find_elements('css selector', 'body')[0]  # 웹페이지 중앙(body) 선택
    body.send_keys(Keys.PAGE_DOWN * 8)  
    time.sleep(3)
    
    soup = BeautifulSoup(browser.page_source, "html.parser")
    item_list = soup.select('div.search_result_list > div')
    
    current_length = len(item_list)
    print(current_length)
    
    if current_length == total:
        break
    
    if current_length == prev_length:
        unchanged_count += 1
    else:
        unchanged_count = 0
    
    if unchanged_count >= 5:
        break
    
    prev_length = current_length
```

```python
results = []
rank = 1

soup = BeautifulSoup(browser.page_source, "html.parser")
item_list = soup.select('div.search_result_list > div')

for item in item_list:

    brand_list = item.select('p.product_info_brand.brand')[0].text.replace('\n','').strip()
    raw_title = item.select('div.product_info_product_name > p.name')[0].text.replace('\n','').strip()
    img_list = item.select('div > picture.picture.product_img > img')[0]['src']
    link_list = "https://kream.co.kr" + item.select('div > a')[0]['href']
    price_list = item.select('div.price.price_area > p.amount')[0].text.replace('\n','').strip()


    if raw_title.startswith('('):
        title_list = raw_title.split(')', 1)[1].strip()
    elif raw_title.startswith('['):
        title_list = raw_title.split(']', 1)[1].strip()
    elif '배송' in raw_title:
        title_list = raw_title.split('송', 1)[1].strip()
    else:
        title_list = raw_title
    data = [rank, brand_list, title_list ,price_list ,price_list ,link_list ,img_list]
    results.append(data)
    rank += 1 


    # 결과 출력 또는 저장
print(results)
```

```python
import pandas as pd
df = pd.DataFrame(results)
df.columns = ['랭킹', '브랜드', '상품명', '기존가격', '현재가격', '사이트' ,'이미지링크']
df.to_excel('./data/KREAM_outer.xlsx',index=False)
```
